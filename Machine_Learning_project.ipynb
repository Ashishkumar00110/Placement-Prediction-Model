# Outlier Analysis

```python
sns.boxplot(y="degree_p", data=df)
```

```python
percentile25=df['degree_p'].quantile(0.25)
percentile75=df['degree_p'].quantile(0.75)
iqr=percentile75-percentile25
```

```python
upper_limit=percentile75+(1.5*iqr)
lower_limit=percentile25-(1.5*iqr)
```

```python
df=df[df['degree_p']<upper_limit]
df=df[df['degree_p']>lower_limit]
```

```python
sns.boxplot(y="degree_p", data=df)
```

```python
col=['ssc_p','hsc_p','degree_p','etest_p','mba_p']
```

```python
for c in col:
  plt.figure()
  sns.boxplot(y=c,data=df)
```

```python
for c in col:
  percentile25=df[c].quantile(0.25)
  percentile75=df[c].quantile(0.75)
  iqr=percentile75-percentile25
  upper_limit=percentile75+(1.5*iqr)
  lower_limit=percentile25-(1.5*iqr)
  df=df[df[c]<upper_limit]
  df=df[df[c]>lower_limit]
  plt.figure()
  sns.boxplot(y=c,data=df)
```

```python
new_df
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
```

```python
X=sc.fit_transform(new_df)
```

```python
X
```

```python
from sklearn.decomposition import PCA
```

```python
pca=PCA(n_components=5)
```

```python
principal_components=pca.fit_transform(X)
```

```python
pca.explained_variance_ratio_.sum()
```

```python
pca_df=pd.DataFrame(data=principal_components,columns=['PC1','PC2','PC3','PC4','PC5'])
```

```python
final_df=pd.concat([pca_df, df.reset_index()[['status','salary']]],axis=1)
```
